```python
# calculate the percentage of missing null values of each columms
def get_null_data(data):
    
    total_rows = data.shape[0]
    for columns in data.columns:
        null_values = data[columns].isnull().sum()
        if null_values > 0:
            per_null_value = float(null_values)*100/total_rows
            print(f'{columns} column has :  {per_null_value} null values')
```

```python
# Function to detect outliers using z-score
def detect_outliers_zscore(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(x - mean) / std for x in data]
    return np.where(np.abs(z_scores) > threshold)[0]

# Detect outliers in column 'A'
outliers = detect_outliers_zscore(df_encoded1['Customers'])

# Print the outliers
print("Outliers in column 'Customers':", df_encoded1['Customers'].iloc[outliers].tolist())
```

```python
# Calculate IQR
Q1 = df_encoded1['Customers'].quantile(0.25)
Q3 = df_encoded1['Customers'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df_encoded1[(df_encoded1['Customers'] < Q1 - 1.5 * IQR) | (df_encoded1['Customers'] > Q3 + 1.5 * IQR)]

# Print outliers
print("Outliers in column 'Customers':")
print(outliers)
```

```python
# Chi-Square Test
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2, mutual_info_classif
```

```python
col = list(df_encoded1.columns)
categorical_features = []
numerical_features = []

for i in col:
    if len(df_encoded1[i].unique()) >6:
        numerical_features.append(i)
    else:
        categorical_features.append(i)
print('categorical_features :', *categorical_features)
print()
print("numerical_features :", *numerical_features)
```

```python
# Chi-Square Test :
features = df_encoded1.loc[:,categorical_features] # char
target = df_encoded1.loc[:,'Store'] # cat
best_features = SelectKBest(score_func = chi2, k='all')
fit = best_features.fit(features, target)

featureScores = pd.DataFrame(data = fit.scores_, index= list(categorical_features),
                             columns = ['Chi Squared Score'])

plt.subplots(figsize=(5,5))
sns.heatmap(featureScores.sort_values(ascending = False, by = 'Chi Squared Score'),
            annot=True, linewidths = 0.4, linecolor = 'black', fmt = '.2f');
plt.title('Selection of Categorical Features')
plt.show()
```

```python

# Calculate adjusted R-squared
n = x_test.shape[0]  # number of observations
p = x_test.shape[1]  # number of predictors (features)
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f'R-squared: {r2}')
print(f'Adjusted R-squared: {adjusted_r2}')
```
